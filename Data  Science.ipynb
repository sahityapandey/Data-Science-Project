# Dropping Missing value

````python
df_cleaned = df.dropna()
````

````python
df.fillna(value={'Arrival Delay in Minutes': 0}, inplace=True)
df
````

````python
df_cleaned = df_cleaned.drop_duplicates()
df
````

##Data Manipulation


````python
import numpy as np
````

````python
 filtered_df = df_cleaned[df_cleaned['Flight Distance'] > 150]
 filtered_df
````

# Grouping by a category and getting the mean of another column

````python
numeric_cols = df_cleaned.select_dtypes(include=np.number).columns
numeric_cols
````

````python
df_cleaned.groupby('Flight Distance')[numeric_cols].mean()
````
